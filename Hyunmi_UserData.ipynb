```python
#pd.options.display.max_rows = 100

#test = pd.read_csv(current_path+'/ml-100k/u.user',sep='|', header=None)
unames = ['userId', 'age', 'gender', 'occupation', 'zipcode']
users =  pd.read_csv(current_path+'/ml-100k/u.user', sep='|', header = None, names = unames)
ratings = pd.read_csv(current_path+'/ml-latest/ratings.csv')
movies = pd.read_csv(current_path+'/ml-latest/movies.csv')
```

```python
# merge dataframes
data = pd.merge(pd.merge(users,ratings),movies)
```

```python
data.groupby('gender')[['rating']].mean()
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc=['mean'], fill_value=0)
mean_ratings
```

```python
# 평점이 200개 이상인 영화만 필터링
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 200]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
#list(mean_ratings)
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by =[('mean','F')],ascending = False)
top_female_ratings[:10]

# list(mean_ratings.columns)
# mean_ratings[('mean','F')]
```

```python
mean_ratings[('mean', 'Diff')] = mean_ratings[('mean', 'F')]-mean_ratings[('mean','M')]
mean_ratings.head()
```

```python
# 남녀 간의 평균 평점 차이가 가장 큰 영화 순위
mean_ratings.sort_values(by=[('mean', 'Diff')], ascending=False)
```

```python
# 성별에 관계없이 표준편차가 큰 영화 -> 성별에 관계없이 호불호가 갈리는 영화
# genre 별로 분류해서 보이면 좋을듯?
ratings_by_std = data.groupby('title')['rating'].std()
ratings_by_std = ratings_by_std.loc[active_titles]
ratings_by_std.sort_values(ascending=False)
```

```python
ratings_std_by_title = data.groupby('title')['rating'].std()
ratings_std_by_title = ratings_std_by_title.loc[active_titles]
ratings_std_by_title.sort_values(ascending=False)[:20]
```

```python
users.drop(axis=1, columns='Age Group',inplace=True)
```

```python
# Continuous Age into Age Group Categories
# add a new column category
# Up to 19 = 10대, 20s = 20대, 30s = 30대, 40-50s = 4-50대, Above 60 = 60세 이상
category = pd.cut(users.age, bins=[0,19,29,39,59, 110], labels =['Up to 19', '20s', '30s', '4-50s', 'Above 60'])
users.insert(2,'Age Group', category)
users.head()
```

```python
# 원래 2-30대로 분류하려고 했었는데 2-30대의 비중이 거의 50퍼센트라서 20대, 30대로 나눔.
users['Age Group'].value_counts(normalize=True)
users['Age Group'].value_counts()

#category.unique()
xlabel = list(users['Age Group'].value_counts().index)
    
age_pie = plt.figure()
ax = age_pie.add_axes([0,0,1,1]) # add_axes(rect) -> rect=[x0, y0, width, height]
ax.pie(users['Age Group'].value_counts(), labels=xlabel, autopct='%1.2f%%')
plt.show()
```

```python
# 수정: datetime.utfcfromtimestamp() 사용
def timetodate():
    file = pd.read_csv(current_path+'/ml-latest-small/ratings.csv')
    frame = pd.DataFrame(file)

    # print(file)
    # print(frame)
    date_data = []
    year_data = []
    month_data = []
    time_data = []
    hour_data = []
    obj = frame.loc[0:, "timestamp"]
    for i in obj:
        timetodate = datetime.utcfromtimestamp(i).strftime('%Y-%m-%d %H:%M:%S')
        dt = datetime.strptime(timetodate, '%Y-%m-%d %H:%M:%S')
        
        # 연 월 일 pd.to_datetime(df['Date'])
        date_data.append(timetodate)
        # Yearly
        year_data.append(dt.year)
        # Monthly
        month_data.append(dt.month)
        # Hourly
        hour_data.append(dt.hour)

    date_series = pd.Series(date_data)
    hour_series = pd.Series(hour_data)
    year_series = pd.Series(year_data)
    month_series = pd.Series(month_data)
    
    frame['date'] = date_series
    frame['year'] = year_series
    frame['month'] = month_series
    frame['hour'] = hour_series
    
    frame.to_csv(current_path+"/data/saved_rating.csv", index=False)
    #print(obj)
    # date = datetime.datetime.fromtimestamp(964982703/1000)

    # print(date)
    # print(date.ctime())
    # print(date.isoformat())
    print("끝")
```
